```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("student_dataset.csv")
```

```python
print(data.isnull().sum())
print("No missing values found")
```

```python
data.drop_duplicates().sum()
#duplicates removed
```

```python
placement_encoder=LabelEncoder()
data["placement_status"]=placement_encoder.fit_transform(data["placement_status"])
```

```python
x=data.drop("placement_status",axis=1)
y=data[["placement_status"]]
```

```python
x_scaler=StandardScaler()
y_scaler=StandardScaler()
x_scaled=x_scaler.fit_transform(x)
y_scaled=y_scaler.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
error=mean_squared_error(y_test,y_predict)
print("Error=",error)
print("RMS=",np.sqrt(error))
```

```python
joblib.dump(model,"placement.pkl")
joblib.dump(x_scaler,"x_scaler.pkl")
joblib.dump(y_scaler,"y_scaker.pkl")
```
