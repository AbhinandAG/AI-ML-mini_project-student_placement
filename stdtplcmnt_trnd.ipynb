```python
import joblib
import pandas as pd
import numpy as np
```

```python
model=joblib.load("placement.pkl")
x_scaler=joblib.load("x_scaler.pkl")
y_scaler=joblib.load("y_scaker.pkl")
```

```python
user_input=pd.DataFrame([[5,73,7,2,18,37,93.27]],columns=['study_hours','attendance','sleep_hours','internet_usage','assignments_completed','previous_score','exam_score'])
userinputscaled=x_scaler.transform(user_input)
prediction_scaled=model.predict(userinputscaled)
prediction=y_scaler.inverse_transform(prediction_scaled)
print(prediction)
```

```python
#Slight error in predicted output
#rounding off
rounded=prediction.astype(np.int32)
rounded
```

```python
if rounded==1:
  print("The student got placed")
else:
  print("The student is not placed")
```

```python
if round(prediction.item()) == 1:
    print("The student got placed")
else:
    print("The student is not placed")
```
